### **Assessing Data**

```python
#Menambahkan kolom Kategori Produk
TokpedFaceWash_dataset_df['Kategori Produk'] = 'Face Wash'
```

```python
#Menilai Data TokpedFaceWash_dataset_df
TokpedFaceWash_dataset_df.info()

#missing values TokpedFaceWash_dataset_df
TokpedFaceWash_dataset_df.isna().sum()
```

```python
#duplikasi TokpedFaceWash_dataset_df
print("Jumlah duplikasi: ", TokpedFaceWash_dataset_df.duplicated().sum())

# Mengecek jumlah duplikasi berdasarkan kolom Nama Produk
print("Jumlah duplikasi berdasarkan Nama Produk: ", TokpedFaceWash_dataset_df.duplicated(subset='Nama Produk').sum())
```

### **Cleaning Data**

```python
# Mengubah tipe data pada colum Harga Produk menjadi Float
TokpedFaceWash_dataset_df['Harga Produk'] = (
    TokpedFaceWash_dataset_df['Harga Produk']
    .str.replace('Rp', '', regex=False)  # Menghapus simbol Rp
    .str.replace('.', '', regex=False)  # Menghapus tanda pemisah ribuan
    .str.replace('No price', '0', regex=False)
    .astype(float)  # Mengonversi ke float
)

# Menampilkan hasil
TokpedFaceWash_dataset_df.info()
print(TokpedFaceWash_dataset_df)
```

```python
#Menangani missing value pada tabel TokpedFaceWash_dataset_df Rating Produk & Terjual
TokpedFaceWash_dataset_df[TokpedFaceWash_dataset_df['Rating Produk'].isna()]
```

```python
#Melakukan metode drooping messing pada tabel TokpedFaceWash_dataset_df
TokpedFaceWash_dataset_df = TokpedFaceWash_dataset_df.dropna(subset=['Rating Produk'])
TokpedFaceWash_dataset_df.info()
```

```python
# Menghapus duplikasi berdasarkan semua kolom dan berdasarkan nama produk
TokpedFaceWash_dataset_df = TokpedFaceWash_dataset_df.drop_duplicates()
print("Jumlah duplikasi: ", TokpedFaceWash_dataset_df.duplicated().sum())

TokpedFaceWash_dataset_df = TokpedFaceWash_dataset_df.drop_duplicates(subset='Nama Produk')
print("Jumlah duplikasi berdasarkan Nama Produk: ", TokpedFaceWash_dataset_df.duplicated(subset='Nama Produk').sum())
```

```python
# Mengecek jumlah baris di mana Harga Produk sama dengan 0
jumlah_baris_harga_0 = len(TokpedFaceWash_dataset_df[TokpedFaceWash_dataset_df['Harga Produk'] == 0])
print("Jumlah baris harga produk sama dengan 0: ", jumlah_baris_harga_0)

# Menghapus baris di mana Harga Produk sama dengan 0
TokpedFaceWash_dataset_df = TokpedFaceWash_dataset_df[TokpedFaceWash_dataset_df['Harga Produk'] != 0]

# Mengecek jumlah baris setelah penghapusan
print("Jumlah baris setelah penghapusan:", len(TokpedFaceWash_dataset_df))
```

```python
#Mengganti tipe data dan nilai colum Terjual menjadi integer
def convert_terjual(value):
    value = value.lower()  # Ubah ke huruf kecil untuk konsistensi
    if 'rb' in value:
        # Mengganti 'rb' dengan '' dan menambahkan 1000 (karena 1rb = 1000)
        return int(value.replace('rb', '').replace('+', '').replace(' terjual', '').strip()) * 1000
    elif ' terjual' in value:
        # Menghapus teks ' terjual' dan mengonversi ke integer
        cleaned_value = value.replace(' terjual', '').strip()
        # Menghapus tanda '+' jika ada
        cleaned_value = cleaned_value.replace('+', '')
        return int(cleaned_value)
    elif value.isdigit():
        # Jika hanya angka saja (misalnya "3 terjual"), kembalikan sebagai integer
        return int(value)
    else:
        return 0  # Jika tidak ada angka yang valid

# Menerapkan fungsi ke kolom 'Terjual'
TokpedFaceWash_dataset_df['Terjual'] = TokpedFaceWash_dataset_df['Terjual'].apply(convert_terjual)

# Menampilkan hasil
print(TokpedFaceWash_dataset_df)
```

```python
TokpedFaceWash_dataset_df.info()
TokpedFaceWash_dataset_df.describe()
```
